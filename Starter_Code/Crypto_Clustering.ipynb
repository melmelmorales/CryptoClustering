```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
df_market_data

market_data_scaled = StandardScaler().fit_transform(
    df_market_data[["price_change_percentage_24h", 
                    "price_change_percentage_7d", 
                    "price_change_percentage_14d",
                    "price_change_percentage_30d",
                    "price_change_percentage_60d",
                    "price_change_percentage_200d",
                    "price_change_percentage_1y"]]
)
```

```python
# Create a DataFrame with the scaled data

df_market_data_scaled = pd.DataFrame(
    market_data_scaled,
    columns=["price_change_percentage_24h", 
                    "price_change_percentage_7d", 
                    "price_change_percentage_14d",
                    "price_change_percentage_30d",
                    "price_change_percentage_60d",
                    "price_change_percentage_200d",
                    "price_change_percentage_1y"]
)

# Copy the crypto names from the original data

df_market_data_scaled["coin_id"] = df_market_data.index

# Set the coinid column as index
df_market_data_scaled = df_market_data_scaled.set_index("coin_id")

# Display sample data
df_market_data_scaled.head(10)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

for i in k:
    model = KMeans(n_clusters=i, random_state=0)
    model.fit(df_market_data_scaled)
    inertia.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data_market = {
    "k": k,
    "inertia": inertia
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow_market = pd.DataFrame(elbow_data_market)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.

elbow_plot_market = df_elbow_market.hvplot.line(x="k", y="inertia", title="Elbow Curve Using Market Data", xticks=k)
elbow_plot_market
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** Based on the elbow curve, it looks like k=4 is the correct one

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4)
```

```python
# Fit the K-Means model using the scaled data
model.fit(df_market_data_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
market_clusters = model.predict(df_market_data_scaled)

# Print the resulting array of cluster values.
print(market_clusters)
```

```python
# Create a copy of the DataFrame
df = df_market_data_scaled.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df["MarketCluster"] = market_clusters

# Display sample data
df.head(10)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

df.hvplot.scatter(
    x="price_change_percentage_24h",
    y= "price_change_percentage_7d",
    by="MarketCluster",
    hover_cols = ["coin_id"], 
    title = "Scatter Plot by Market Segment - k=4"
)
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
market_pca_data = pca.fit_transform(df_market_data_scaled)

# View the first five rows of the DataFrame. 
market_pca_data[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** About 90% of the total variance is condensed into the 3 PCA variables

```python
# Create a new DataFrame with the PCA data.

df_market_pca = pd.DataFrame(market_pca_data, columns =["PC1", "PC2", "PC3"])

# Copy the crypto names from the original data

df_market_pca["coin_id"] = df_market_data.index

# Set the coinid column as index

df_market_pca = df_market_pca.set_index("coin_id")

# Display sample data
df_market_pca.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values

inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list

for i in k:
    model = KMeans(n_clusters=i, random_state=0)
    model.fit(df_market_pca)
    inertia.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve

elbow_data_market_pca = {
    "k": k,
    "inertia": inertia
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow_market_pca = pd.DataFrame(elbow_data_market_pca)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.

elbow_plot_market_pca = df_elbow_market_pca.hvplot.line(x="k", y="inertia", title="Elbow Curve Using Market PCA Data", xticks=k)
elbow_plot_market_pca
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** k = 4


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No, it does not differ

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model_pca = KMeans(n_clusters=4)
```

```python
# Fit the K-Means model using the PCA data
model_pca.fit(df_market_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
market_clusters_pca = model_pca.predict(df_market_pca)

# Print the resulting array of cluster values.
print(market_clusters_pca)
```

```python
# Create a copy of the DataFrame with the PCA data
df_pca = df_market_pca.copy()

# Add a new column to the DataFrame with the predicted clusters
df_pca["MarketClusterPCA"] = market_clusters_pca

# Display sample data
df_pca.head(10)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

df_pca.hvplot.scatter(
    x="PC1",
    y= "PC2",
    by="MarketClusterPCA",
    hover_cols = ["coin_id"], 
    title = "Scatter Plot by Market Segment (PCA Data) - k=4"
)
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the Elbow curves

elbow_plot_market = df_elbow_market.hvplot.line(x="k", y="inertia", title="Elbow Curve Using Market Data", xticks=k, color='blue', label='Market Data')
elbow_plot_market_pca = df_elbow_market_pca.hvplot.line(x="k", y="inertia", title="Elbow Curve Using Market PCA Data", xticks=k, color='red', label='Market PCA Data')

# Combine the two plots
composite_plot = elbow_plot_market * elbow_plot_market_pca

# Display the composite plot
composite_plot
```

```python
# Composite plot to contrast the clusters

# Scatter plot for the original market data
scatter_plot_market = df.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="MarketCluster",
    hover_cols=["coin_id"],
    title="Scatter Plot by Market Segment - k=4",
    cmap='viridis'  # You can choose a colormap that fits your preference
)

# Scatter plot for the PCA-transformed market data
scatter_plot_market_pca = df_pca.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="MarketClusterPCA",
    hover_cols=["coin_id"],
    title="Scatter Plot by Market Segment (PCA Data) - k=4",
    cmap='plasma'  # Different colormap to distinguish between the plots
)

# Combine the two plots into a composite plot
composite_scatter_plot = scatter_plot_market * scatter_plot_market_pca

# Display the composite plot
composite_scatter_plot
```
